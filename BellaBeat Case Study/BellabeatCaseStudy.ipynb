```python
df['ActivityDate']=pd.to_datetime(df['ActivityDate'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Basic user activity analysis
print("Total number of unique users:", df['Id'].nunique())
print("Date range of data:", df['ActivityDate'].min(), "to", df['ActivityDate'].max())
```

```python
# 2. Average daily activity metrics
avg_steps = df['TotalSteps'].mean()
avg_distance = df['TotalDistance'].mean()
avg_calories = df['Calories'].mean()


print(f"Average daily steps: {avg_steps:.0f}")
print(f"Average daily distance: {avg_distance:.2f} km")
print(f"Average daily calories burned: {avg_calories:.0f}")
```

```python
plt.figure(figsize=(15, 10))  

plt.subplot(1, 3, 1)
plt.hist(df['TotalSteps'], bins=30, color='LightGreen', alpha=0.7, edgecolor='black', linewidth=0.5)
plt.title('Distribution of Daily Steps', fontsize=16, fontweight='bold')
plt.xlabel('Total Steps', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adjust layout to prevent overlapping
plt.tight_layout(pad=3.0)
plt.show()
```

```python
plt.figure(figsize=(12, 8))  
plt.scatter(df['TotalSteps'], df['Calories'], alpha=0.6, color='orange', s=50)  
plt.title('Steps vs Calories Burned', fontsize=16)
plt.xlabel("Total Steps", fontsize=14)
plt.ylabel("Calories", fontsize=14)
plt.grid(True, alpha=0.3) 
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(20, 8)) 

plt.subplot(1, 3, 3)
daily_avg = df.groupby('ActivityDate')['TotalSteps'].mean()
plt.plot(daily_avg.index, daily_avg.values, color='green', linewidth=2.5, marker='o', markersize=4)
plt.title('Average Daily Steps Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Steps', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')

# Format the plot for better readability
plt.gcf().autofmt_xdate()  # Auto-format dates for better fit

plt.tight_layout(pad=3.0)
plt.show()
```

```python
# 1. Categorize users by activity level
def categorize_activity_level(steps):
    if steps < 5000:
        return 'Sedentary'
    elif steps < 7500:
        return 'Lightly Active'
    elif steps < 10000:
        return 'Fairly Active'
    else:
        return 'Very Active'

# Apply categorization
df['ActivityLevel'] = df['TotalSteps'].apply(categorize_activity_level)
```

```python
# 2. Analyze user activity patterns
user_summary = df.groupby('Id').agg({
    'TotalSteps': 'mean',
    'TotalDistance': 'mean', 
    'Calories': 'mean',
    'VeryActiveMinutes': 'mean',
    'FairlyActiveMinutes': 'mean',
    'LightlyActiveMinutes': 'mean',
    'SedentaryMinutes': 'mean'
}).round(2)
```

```python
# Activity level distribution
activity_distribution = df['ActivityLevel'].value_counts()
print("Activity Level Distribution:")
print(activity_distribution)
```

```python
 #Day of week analysis
df['DayOfWeek'] = df['ActivityDate'].dt.day_name()
weekly_activity = df.groupby('DayOfWeek')['TotalSteps'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
```

```python
# 4. Create insights visualizations
plt.figure(figsize=(15, 10))
```

```python
# Plot 1: Activity level pie chart
plt.subplot(2, 3, 1)
plt.pie(activity_distribution.values, labels=activity_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('User Activity Level Distribution')
```

```python
#Plot 2: Weekly activity pattern
plt.figure(figsize=(13, 9))  

plt.subplot(2, 3, 2)
plt.bar(weekly_activity.index, weekly_activity.values, color='lightcoral', edgecolor='darkred', linewidth=1.2, alpha=0.8)
plt.title('Average Steps by Day of Week', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, fontsize=12, ha='right')
plt.ylabel('Average Steps', fontsize=14)
plt.yticks(fontsize=12)
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Add value labels on top of bars
for i, v in enumerate(weekly_activity.values):
    plt.text(i, v + max(weekly_activity.values)*0.01, f'{v:.0f}', 
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout(pad=3.0)
plt.show()
```

```python

plt.figure(figsize=(16, 12))  # Larger figure for better visibility

# Plot 3: Active minutes correlation
plt.subplot(2, 3, 3)
plt.scatter(df['VeryActiveMinutes'], df['Calories'], alpha=0.7, color='red', s=60, edgecolor='darkred', linewidth=0.5)
plt.title('Very Active Minutes vs Calories', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Very Active Minutes', fontsize=14)
plt.ylabel('Calories', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
```

```python
#Plot 4: Sedentary vs Active time
plt.subplot(2, 3, 4)
plt.scatter(df['SedentaryMinutes'], df['TotalSteps'], alpha=0.6, color='purple')
plt.title('Sedentary Time vs Total Steps')
plt.xlabel('Sedentary Minutes')
plt.ylabel('Total Steps')
plt.tight_layout()
plt.show()
```

```python
#key insights
print(f"\nKey Insights:")
print(f"Most active day: {weekly_activity.idxmax()}")
print(f"Least active day: {weekly_activity.idxmin()}")
print(f"Average very active minutes: {df['VeryActiveMinutes'].mean():.1f}")
print(f"Average sedentary hours: {df['SedentaryMinutes'].mean()/60:.1f}")
```
